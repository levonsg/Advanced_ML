```python
def silhouette(y_pred, x):
    n, count = np.unique(y_pred, return_counts==True)
    n= len(n)
    a = np.zeros(len(x))
    k = len(x)
    b  = np.zeros((k,n))
    for i in range(len(y_pred)):
        for j in range(len(y_pred)):
            if (i != j) and (y_pred[i] == y_pred[j]) :
                a[i] += euclidean_distances(x[i], x[j])
            elif i != j and y_pred[i] != y_pred[j]:
                b[i][y_pred[j]] += euclidean_distances(x[i], x[j])
                
        a[i] += a[i] / (len(y_pred[y_pred == y_pred[i]]-1))      
        b[i]  = b[i]/count
        
        b = np.sort(b, axis=1)
        b = b[:,1]
        s =  (b-a)/(np.maximum(a,b))
    
    
    return s.mean()
```
