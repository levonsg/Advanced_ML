# Decision Tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

details - https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)
```

```python
data = pd.read_csv('./classification.csv')


X_train = data.drop(columns='default', )
X_train = X_train.astype(float)
y_train = data['default']
X_train.values
```

```python
clf = DecisionTreeClassifier(class_weight='balanced', max_leaf_nodes=10, random_state=10)
clf.fit(y_train.values, y_train.values)
print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))

# Can we guess what would this scores be?
```

### Messing with train data

```python
import numpy as np

X_train = np.append(X_train, [X_train[0]], axis=0)
y_train = np.append(y_train, [not y_train[0]], axis=0)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))

# what would the scores be now?
```

```python
import random

for _ in range(3):
    X_train = np.append(X_train, [X_train[0]], axis=0)
    y_train = np.append(y_train, random.sample([0,1,2], 1), axis=0)
```

```python
clf = DecisionTreeClassifier(class_weight='balanced')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))
```

# Visualizing DT

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plot_tree(clf, filled=True)
plt.show()
```
