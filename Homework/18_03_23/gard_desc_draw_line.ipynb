```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, iterations=100):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Initialize the weights and bias to zero
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        y = np.squeeze(y)
        for i in range(self.iterations):
            y_predicted = np.dot(X, self.weights) + self.bias
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
            plt.scatter(X, y)
    
            x_range = np.linspace(np.min(X), np.max(X), 100)
            y_range = self.bias + self.weights*x_range
            plt.plot(x_range, y_range, c='r', linewidth=0.4)
    
            plt.xlabel("Length of Membership")
            plt.ylabel("Length of Membership")
            plt.title("Regression line \"Length of Membership\" vs \"Yearly Amount Spent\" ")
    
    
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
df = pd.read_csv('Ecommerce.csv')
```

```python
X = df.loc[:, df.columns == "Length of Membership"]
X
```

```python
y = df.loc[:, df.columns == "Yearly Amount Spent"]
y
```

```python


lr = LinearRegression(learning_rate=0.02, iterations=400)
lr.fit(X, y)


# y_pred = lr.predict(X)
# print(y_pred)  # Output: [3.993, 5.004, 6.015]
```
