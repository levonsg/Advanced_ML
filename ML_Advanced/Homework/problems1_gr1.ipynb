# 2 Find unique values of the given numpy array.

```python
import numpy as np
```

```python
ar = np.array([1, 2, 3, 4, 1, 2, 1])
unique = []
for item in ar:
    new_ar = np.delete(ar, 0)
    ar = new_ar
    if (item not in new_ar) and (item not in unique):
        unique.append(item)
print(unique)
```

# 1 Generate random integer numbers of a given shape. From random generators you can use only np.random.rand()

```python
def int_gen(start, end, n):
    int_ar = (end - start) * np.random.rand(n) + start
    
    return int_ar.astype(int)
    
    
```

# 5 Replace random values (~10 % of values) of car data.csv with np.nan; Then fill that nan values with most common values of the column for categorical columns and with mean values from numerical columns.

```python
import pandas as pd
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
nan_size = int(len(df) * 0.1)
nan_size
p =nan_size/len(df)
p
```

```python
#mask = np.random.choice([True, False], size=len(df), p=[0.1, 0.9])

for col in df.columns:
    mask = np.random.choice([True, False], size=len(df), p=[0.1, 0.9])
    df.loc[mask, col] = np.NaN
```

```python
df.head(50)
```

```python
for col in df.columns:
    if df[col].dtype=='object':
        df[col] = df[col].fillna(df[col].value_counts().idxmax())
    else:
        df[col] = df[col].fillna(df[col].mean())
df['Year'] = df['Year'].astype(int)
```

# 6 Plot graphics of functions x^2 and sin(x) in the same figure. Make one of them regular plot and the other one with dashed lines.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 1000)
y1 = x**2
y2 = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y1, label='x^2')
ax.plot(x, y2, '--', label='sin(x)')
ax.legend()

plt.show()
```

# 7 Scatter plot data points from car data.csv using selling price and present price as data point coordinates. Differentiate the points by colors, each color corresponds to a unique fuel type (e.g., all Petrols are blue and all Diesels are red). Add a legend for each color.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('car data.csv')
colors = {'Petrol': 'blue', 'Diesel': 'red', 'CNG': 'green', 'LPG': 'orange'}
fig, ax = plt.subplots()
for fuel_type, color in colors.items():
    data = df[df['Fuel_Type'] == fuel_type]
    ax.scatter(data['Selling_Price'], data['Present_Price'], color=color, label=fuel_type)
ax.legend()
ax.set_xlabel('Selling Price')
ax.set_ylabel('Present Price')
plt.show()
```

# 3 Find value of derivative of a function of 1 variable in given point. The function can be only polynomial and is given by string. For example func('3x^2+4x+4', 3) == 22 (for your ease - lets limit coefficients to be only integers)

```python
 import re
def func(poly, x):
    # Split the polynomial string into its individual terms
    terms_ = poly.replace('-', '+-')
    terms = terms_.split('+')
    derivative = 0
    for term in terms:
        if 'x^' in term:
            coef, power = term.split('x^')
            
            power = int(power)
            if power == 0:
                derivative += 0
            else:
                if coef == '':
                    coef = 1
                elif coef == '-':
                    coef = -1
                derivative += int(coef) * power * x**(power - 1)
        elif 'x' in term:
            
            coef, _ = term.split('x')
            if coef == '':
                coef = 1
            elif coef == '-':
                coef = -1
            derivative += int(coef)

    return derivative

```
